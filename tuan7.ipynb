```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sas

data = pd.read_csv('50_Startups.csv')
data.head()



```

```python
sas.pairplot(data)
```

```python
corr = data.corr()
sas.heatmap(corr,annot=True)
```

```python
X = data.drop('Profit',axis=True)
X = X.drop('State',axis = True)

y = data['Profit']
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X = SC.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)

X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#hoc
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test).round(1)

calculation = pd.DataFrame(np.c_[y_test,y_pred],columns = ['Original Profit','Predicted Profit'])
```

```python
regressor.score(X_test,y_test)
```

```python
X=pd.DataFrame(data['TV'])
y= data['sales']

model.score(X,y)
y_pred = model.predict(X)
residual = data['sales']-y_pred
residual.describe()

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X =sc.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
X_train.shape


from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)

y_pred = regressor.predict(X_test).round(1)
calculation = pd.DataFrame(np.c_[y_test,y_pred], columns=['Original sales','Predicted sales'])

#Do su dung bac 1 nen do chinh xac thap
print('Độ chính xác của mô hình')
print(regressor.score(X_train,y_train))
print(regressor.score(X_test,y_test))
```
